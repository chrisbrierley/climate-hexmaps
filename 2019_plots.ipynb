```python
import xarray as xr
import numpy as np

import matplotlib.pyplot as plt      # for plotting
import cartopy.crs as ccrs           # for adding maps to the plots
```

```python
# Load the data into a variable called dat

# you can load several files by using a wildchard character * with xr.open_mfdataset
ds = xr.open_mfdataset("/data/met/hadUK-grid/dmo/rainfall_hadukgrid_uk_12km_day_18910101-20211231.nc")

# display the contents of the data
ds
```

```python
# the dataset 'ds' contains multiple variables - we only want to work with the variable 'tasmin', so extract it using 
rainfall = ds["rainfall"]
```

```python
ann_mean = rainfall.resample(time = "1AS-DEC",).mean()
ann_mean.time
```

```python
ann_mean.sel(time = "20191201").plot()
```

```python
# set the figure size and resolution, along with foreground and edge colours (particularly important when saving, as the default is a transparent background)
plt.figure(figsize = (5,7), dpi = 250, facecolor='w', edgecolor='k')

# tell matplotlib which projection you're using, and draw the coastlines
ax = plt.axes(projection = ccrs.OSGB(approx = False))
ax.coastlines(resolution = '10m')

# add the plot to the axes we just created
# don't draw the colour scale, but save the information and plot it manually at the bottom of the image
cbar = ann_mean.sel(time = "20191201").plot(ax = ax, cmap = "Blues", add_colorbar = False)
plt.colorbar(cbar, location = "right", pad = 0.05, label = "Average Daily Precipitation (mm/day)")

# add x and y axes
# matplotlib can't add and label gridlines automatically for OSGB coordinate system - so will create manually
xlines = np.arange(0,8e5,2e5)
ylines = np.arange(0,14e5,2e5)





# save the map as a .png
plt.savefig("./example_rainfall-plot.png", bbox_inches = "tight")
# plt.close()      # if you just want to save the plot without displaying it, uncomment this line
```
